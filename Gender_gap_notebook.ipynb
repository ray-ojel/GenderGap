# Visualizing the Gender Gap in College Degrees

[The Department of Education Statistics](https://nces.ed.gov/programs/digest/2013menu_tables.asp) releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. 

The data set is broken up into 17 categories of degrees, with each column as a separate category.

Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. You can download the dataset Randal compiled [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv). 

Questions:
* Is there a noticeable gap between the number of male and female graduates in a given major?
* What's the smallest/largest gap majors?

Conclusion:
* In STEM majors, the biggest gap is in Engineering and the smallest gap is in Math & Stats
* In liberal arts majors, the gap is small overall but the biggest is in the English majors and the smallest is in Social Sciences and History
* In other majors, Health Professions, Public Admin, and Education have a big gap and is filled by majority women, Business and Agriculture seem to have closed the gap. 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

Now we're going to visualize the data of STEM majors, liberal arts majors and other majors.

To be mindful of colorblind readers, we use the tableu palette with matplotlib to import the blue, orange and grey from the colorbind-10 with their RGB values, and since matplotlib only accepts values from 0 to 1 for RGB, we have to take the ratios over 255 (RGB Range from 0 to 255).

Then, to improve our data-ink ratio, we remove the tick marks, remove the spines of the graph and make the lines thicker for easier data visualization. 

We also remove the x ticks for middle plots and shorten y to only 0, 50, and 100. 

Since we're combining in one figure, I'll comment on the first subplot and the rest are exactly the same.

```python
# set colors to variables that matplotlib would accept
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_grey = (171/255, 171/255, 171/255)
```

```python
# import STEM, liberal arts, and other majors to their respective lists
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 
             'Physical Sciences', 'Computer Science', 'Engineering']

lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism',
                 'Art and Performance', 'Social Sciences and History']

other_cats = ['Health Professions', 'Public Administration', 'Education', 
              'Agriculture','Business', 'Architecture']
```

```python
# create a new figure with dimensions 16 and 20
fig = plt.figure(figsize = (16, 20))

# create the first subplot for STEM majors
for i in range(0, 6):
    
    # position of the subplots in the figure
    cat_index = [1, 4, 7, 10, 13, 16]
    
    # create a new subplot at each position
    ax = fig.add_subplot(6, 3, cat_index[i])
    
    # create two lines in the same plot, one for women and another for men 
    # we use 100 - Series for men because values are listed as percents of women
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[i]], 
            c=cb_dark_blue, label = 'Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[i]], 
            c=cb_orange, label = 'Men', linewidth=3)
    
    # customize appearance of x and y axes
    # improve data-ink ratio
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[i])
    ax.set_yticks([0, 50, 100])
    
    # draw a grey horizontal line at y = 50
    ax.axhline(50, color = cb_grey, alpha=0.5)
    
    # delete ticks around the plot
    ax.tick_params(bottom="off", top="off", 
                   left="off", right="off", labelbottom = 'off')

    # add annotations to the top and bottommost plots
    if i == 0:
        ax.text(2005, 87, 'Women')
        ax.text(2002, 8, 'Men')
    elif i == 5:
        ax.text(2005, 73, 'Men')
        ax.text(2001, 25, 'Women')
        ax.tick_params(labelbottom = 'on')
        
    # remove the spines all around 
    # ax.spines creates a dictionary
    for key, spine in ax.spines.items():
        spine.set_visible(False)

# create the second subplot for liberal arts majors
# only contains 5 majors so last plot will be empty
for i in range(0, 5):
    cat_index = [2, 5, 8, 11, 14]
    ax = fig.add_subplot(6, 3, cat_index[i])
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[i]], 
            c=cb_dark_blue, label = 'Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[i]], 
            c=cb_orange, label = 'Men', linewidth=3)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[i])
    ax.set_yticks([0, 50, 100])
    ax.axhline(50, color = cb_grey, alpha=0.5)
    ax.tick_params(bottom="off", top="off", 
                   left="off", right="off", labelbottom = 'off')

    if i == 0:
        ax.text(2002, 80, 'Women')
        ax.text(2002, 15, 'Men')
    elif i == 4:
        ax.tick_params(labelbottom = 'on')
        
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
# create a third subplot for other majors        
for i in range(0, 6):
    cat_index = [3, 6, 9, 12, 15, 18]
    ax = fig.add_subplot(6,3, cat_index[i])
    ax.plot(women_degrees['Year'], women_degrees[other_cats[i]],
           c=cb_dark_blue, label = 'Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[i]],
           c=cb_orange, label = 'Men', linewidth=3)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[i])
    ax.set_yticks([0, 50, 100])
    ax.axhline(50, color = cb_grey, alpha=0.5)
    ax.tick_params(bottom="off", top="off", 
                   left="off", right="off", labelbottom = 'off')

    if i == 0:
        ax.text(2002, 95, 'Women')
        ax.text(2002, 5, 'Men')
    elif i == 5:
        ax.text(2001, 65, 'Men')
        ax.text(2001, 30, 'Women')
        ax.tick_params(labelbottom = 'on')
        
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
        
# plt.savefig('gender_degrees.png')        
plt.show()
```
